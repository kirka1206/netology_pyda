# Задание 1

```python
geo_data = {

    'Центр': ['москва', 'тула', 'ярославль'],

    'Северо-Запад': ['петербург', 'псков', 'мурманск'],

    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']

}

keywords_df = pd.read_csv('keywords.csv')
```

```python
def find_geo(keyword):
    for key, value in geo_data.items():
        if keyword.lower() in value:
            return key
    return 'undefined'
```

```python
keywords_df['region'] = keywords_df['keyword'].apply(find_geo)
keywords_df[keywords_df['region'] != 'undefined']
```

# Задание 2

```python
ratings_df = pd.read_csv('ml-latest-small/ratings.csv')
movies_df = pd.read_csv('ml-latest-small/movies.csv')

def get_class_by_rating(rating):
    if rating <= 2:
        return 'низкий рейтинг'
    elif rating <= 4:
        return 'средний рейтинг'
    else:
        return 'высокий рейтинг'
```

```python
mean_ratings = ratings_df.groupby('movieId').mean()[['rating']]
mean_ratings['class'] = mean_ratings['rating'].apply(get_class_by_rating)
```

```python
res = movies_df.join(mean_ratings, how='left', on='movieId').set_index('movieId')
res.head()
```

```python
from datetime import datetime
import numpy as np
ratings_df = pd.read_csv('ml-latest-small/ratings.csv')
ratings_df['timestamp'] = ratings_df['timestamp'].apply(datetime.utcfromtimestamp)
```

# Задание 4

```python
import re

ratings_df = pd.read_csv('ml-latest-small/ratings.csv')
movies_df = pd.read_csv('ml-latest-small/movies.csv')
years = list(map(str, range(1950, 2011)))
```

```python
mean_ratings = ratings_df.groupby('movieId').mean()[['rating']]
res = movies_df.join(mean_ratings, how='left', on='movieId').set_index('movieId')
```

```python
res['year'] = res['title'].str.extract(r'(\d{4})')
res['year'] = res['year'].apply(lambda year: int(year) if year in years else 1900)
```

```python
res.groupby('year')[['rating']].mean().sort_values(by='rating', ascending=False)
```
